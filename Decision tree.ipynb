```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('adult.csv')
dataset.shape
```

```python
dataset.tail()
```

```python
''' confusion matrix is used 
                                            Predicted class
______________________________________________________________________________
                              |                         |
                              |  Class=yes              |             class=no
Actual Class     class=yes    |  True +ve               |             Flase -ve
                 class=no     |  False +ve              |             True -ve
_______________________________________________________________________________                 

TP = True Positive
TN = True Negative
FP = False Positive
FN = False Negative

Accuracy = TP+TN/TP+FP+FN+TN
Precision = TP/TP+FP
Recall = TP/TP+FN
F1 Score = 2 * (Recall * Precision) / (Recall + Precision)
F1 score is more important Accuracy

                      
'''
```

```python
x=dataset[['capital.loss','education.num','hours.per.week']]
y=dataset['income']
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier= DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
df=pd.DataFrame({'Actual': y_test,'Predicted':y_pred,'true':y_test==y_pred})
```

```python
df
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
